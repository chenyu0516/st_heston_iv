```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/test_options_data.csv')
df.head()
```

```python
df_trading = df[df['DTE']==1].reset_index()
df_trading.head()
```

```python
import os, sys
sys.path.append(os.path.abspath("../"))
from src.models.iv_test import bisection_call_iv, bisection_put_iv
from src.models.iv_test import HestonSmallTimeIV
from scipy.optimize import least_squares
param = (-0.4, 0.1, 0.5)
model = HestonSmallTimeIV(param, None, None)
```

```python
nan = np.nan
def residuals(params, x, market_iv):
    rho, sigma, y0 = params
    model_iv = model.IV_curve([rho, sigma, y0], x)
    return model_iv - market_iv

def calibrate_iv_smile(x, market_iv, initial_guess=(0.0, 0.3, 0.04)):

    res = least_squares(residuals, initial_guess, bounds=([-0.999, 1e-6, 1e-6], [0.999, 100.0, 3.0]), args=(x, market_iv))
    rho, sigma, y0 = res.x
    return rho, sigma, y0

rho_list = []
sigma_list = []
y0_list = []
for i in range(df_trading.shape[0]):

    S0 = df_trading.loc[i,'S0']
    K = np.array(eval(df_trading.loc[i,'K']))
    r = df_trading.loc[i,'r']
    T = df_trading.loc[i,'DTE']/365
    call = np.array(eval(df_trading.loc[i,'c_last']))
    put = np.array(eval(df_trading.loc[i,'p_last']))

    call = call[(K>=S0*0.9)&(K<=S0*1.1)]
    put = put[(K>=S0*0.9)&(K<=S0*1.1)]
    K = K[(K>=S0*0.9)&(K<=S0*1.1)]
    put_iv = bisection_put_iv(S0, K, r, T, put)
    call_iv = bisection_call_iv(S0, K, r, T, call)
    
    # OTM option IV
    index1 = (K < S0)
    index2 = (K > S0)
    otm_iv = np.zeros(len(K))

    otm_iv[index1] = put_iv[index1]
    # for put options OTM is strike price <= current option price
    otm_iv[index2] = call_iv[index2]
    # > current option price for call options

    K = K[otm_iv != 0]
    otm_iv = otm_iv[otm_iv != 0]
    # exclude all the OTM implied volatility = 0

    lnK = np.log(K/S0)
    rho, sigma, y0 = calibrate_iv_smile(lnK, otm_iv)
    
    rho_list.append(rho)
    sigma_list.append(sigma)
    y0_list.append(y0)

df_trading['rho'] = rho_list
df_trading['sigma'] = sigma_list
df_trading['y0'] = y0_list
    
    
   
```
