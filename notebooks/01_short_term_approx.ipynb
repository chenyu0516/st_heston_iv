```python
from src.models.iv import HestonSmallTimeIV
import matplotlib.pyplot as plt
import numpy as np

model = HestonSmallTimeIV((-0.4, 0.1, 0.5), None, None)
```

```python
us = np.linspace(-10, 10, 100)
# plt.plot(us, sa_iv_cal.IV_curve(param=[-.04, 0.2, 0.04], lnK=us))
for rho in [0.4, 0, -0.4]:
    plt.plot(us, [model.Lambda([rho, 0.2, 0.2], u) for u in us], label=f"rho={rho}")
plt.title("(x)")
plt.xlabel("ln(K/S0)")
plt.ylabel("(x)")
plt.legend()
plt.grid()
plt.show()

us = np.linspace(-2, 2, 100)
for rho in [0.4, 0, -0.4]:
    plt.plot(us, [model.legendre_transform([rho, 0.2, 0.2], u) for u in us], label=f"rho={rho}")
plt.title("Rate Function")
plt.xlabel("ln(K/S0)")
plt.ylabel("Rate Function")
plt.legend()
plt.grid()
plt.show()

for rho in [0.4, 0, -0.4]:
    plt.plot(us, model.IV_curve([rho, 0.2, 0.2], us), label=f"rho={rho}")
plt.title("Implied Volitility Curve")
plt.xlabel("ln(K/S0)")
plt.ylabel("IV")
plt.legend()
plt.grid()
plt.show()
```
